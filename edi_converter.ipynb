````python
import json
from typing import List, Dict, Any, Optional
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.schema import BaseOutputParser
from langchain_core.runnables import RunnableSequence
from pydantic import BaseModel, Field
````

````python
# Define the output structure
class ExtractedEntity(BaseModel):
    entity: str = Field(description="The entity name")
    value: Optional[str] = Field(description="The extracted value, null if not found")
    required: str = Field(description="Whether the entity is mandatory (M) or optional (O)")
    found: bool = Field(description="Whether the entity was found in the text")

class EntityExtractionResult(BaseModel):
    extracted_entities: List[ExtractedEntity] = Field(description="List of extracted entities")
    confidence_score: float = Field(description="Overall confidence score (0-1)")

# Custom output parser
class EntityOutputParser(BaseOutputParser[EntityExtractionResult]):
    def parse(self, text: str) -> EntityExtractionResult:
        try:
            # Clean the text to extract JSON
            if "```json" in text:
                json_start = text.find("```json") + 7
                json_end = text.find("```", json_start)
                json_text = text[json_start:json_end].strip()
            else:
                json_text = text.strip()
            
            data = json.loads(json_text)
            return EntityExtractionResult(**data)
        except Exception as e:
            # Fallback parsing
            return EntityExtractionResult(
                extracted_entities=[],
                confidence_score=0.0
            )

# Initialize the LLM
llm = ChatOpenAI(
    model_name="gpt-4o",
    temperature=0.1
)

# Create the prompt template
prompt_template_entities_extraction = ChatPromptTemplate.from_template("""
You are an expert EDI (Electronic Data Interchange) data extraction specialist. Your task is to extract specific entities from the given text.

**Instructions:**
1. Carefully analyze the provided text
2. Extract values for each specified entity if present
3. Mark entities as found (true) or not found (false)
4. Maintain the required status (M=Mandatory, O=Optional) as provided
5. Provide a confidence score between 0 and 1 for the overall extraction

**Text to analyze:**
{text}

**Entities to extract:**
{entities}

**Output format (JSON):**
```json
{{
    "extracted_entities": [
        {{
            "entity": "ENTITY_NAME",
            "value": "extracted_value_or_null",
            "required": "M_or_O",
            "found": true_or_false
        }}
    ],
    "confidence_score": 0.95
}}
```

**Extraction Guidelines:**
- DATE: Look for dates in various formats (MM/DD/YYYY, Month DD YYYY, etc.)
- INVOICE NUMBER: Look for invoice numbers, invoice IDs, or similar identifiers
- PURCHASE ORDER NUMBER: Look for PO numbers, purchase order references
- RELEASE NUMBER: Look for release numbers or blanket order releases
- CHANGE ORDER SEQUENCE NUMBER: Look for change order numbers or revisions
- TRANSACTION TYPE CODE: Look for transaction codes, billing codes, or transaction types
- Numbers and codes should be extracted exactly as they appear
- If an entity is not found, set value to null and found to false
- Be precise and only extract what's clearly present in the text

Extract the entities now:
""")
    
# Create the output parser
output_parser = EntityOutputParser()

# Create the chain
chain_entities_extraction = (
        prompt_template_entities_extraction
        | llm
        | output_parser
    )
    

def extract_entities(text: str, entities: List[Dict[str, str]]) -> EntityExtractionResult:
    """
    Extract entities from text using the LLM chain.
    
    Args:
        text: The text to analyze
        entities: List of entities to extract with their required status
    
    Returns:
        EntityExtractionResult with extracted entities and confidence score
    """
    
    # Format entities for the prompt
    entities_str = json.dumps(entities, indent=2)
    
    # Run the chain
    try:
        result = chain_entities_extraction.invoke({'text': text, 'entities': entities_str})
        return result
    except Exception as e:
        print(f"Error during extraction: {e}")
        # Return empty result on error
        return EntityExtractionResult(
            extracted_entities=[
                ExtractedEntity(
                    entity=entity['entity'],
                    value=None,
                    required=entity['required'],
                    found=False
                ) for entity in entities
            ],
            confidence_score=0.0
        )
````

````python
# Example usage
def demo_extraction():
    """Demonstrate the entity extraction chain"""
    
    # Your example data
    text = "This invoice (number INV145789) was issued on June 6 2025 for items or services ordered on purchase order PO987654 (dated June 1 2025). No blanket-order release number or change-order revision applies, and the billing is a debit memo (code DR)."
    text2 = "The Bill-To party (identified by \"BY\") is ABC CORPORATION, with an internal identification code 12345 (qualifier 92)."

    entities = [
        {'entity': 'DATE', 'required': 'M'},
        {'entity': 'INVOICE NUMBER', 'required': 'M'},
        {'entity': 'DATE', 'required': 'O'},
        {'entity': 'PURCHASE ORDER NUMBER', 'required': 'O'},
        {'entity': 'RELEASE NUMBER', 'required': 'O'},
        {'entity': 'CHANGE ORDER SEQUENCE NUMBER', 'required': 'O'},
        {'entity': 'TRANSACTION TYPE CODE', 'required': 'O'},
        {'entity': 'TRANSACTION SET PURPOSE CODE', 'required': 'O'},
        {'entity': 'ACTION CODE', 'required': 'O'},
        {'entity': 'INVOICE NUMBER', 'required': 'O'},
        {'entity': 'HIERARCHICAL STRUCTURE CODE', 'required': 'O'}
    ]
    entities2 = [
        {'entity': 'ENTITY IDENTIFIER CODE', 'required': 'M'},
        {'entity': 'NAME', 'required': 'C'},
        {'entity': 'IDENTIFICATION CODE QUALIFIER', 'required': 'C'},
        {'entity': 'IDENTIFICATION CODE', 'required': 'C'},
        {'entity': 'ENTITY RELATIONSHIP CODE', 'required': 'O'},
        {'entity': 'ENTITY IDENTIFIER CODE', 'required': 'O'}
    ]
    
    # Extract entities
    #result = extract_entities(text, entities)
    result = extract_entities(text2, entities2)

    # Display results
    print("=== ENTITY EXTRACTION RESULTS ===")
    print(f"Confidence Score: {result.confidence_score}")
    print("\nExtracted Entities:")
    
    for entity in result.extracted_entities:
        status = "✓ FOUND" if entity.found else "✗ NOT FOUND"
        req_status = "MANDATORY" if entity.required == "M" else "OPTIONAL"
        print(f"  {entity.entity} ({req_status}): {entity.value} [{status}]")
    
    return result

# Run the demo
if __name__ == "__main__":
    result = demo_extraction()
````

````python
from langchain_core.output_parsers import StrOutputParser


# Custom output parser
class EDIExpressionOutputParser(BaseOutputParser[str]):
    def parse(self, text: str) -> str:
        try:
            # Clean the text to extract JSON
            if "```json" in text:
                json_start = text.find("```json") + 7
                json_end = text.find("```", json_start)
                json_text = text[json_start:json_end].strip()
            else:
                json_text = text.strip()
            
            data = json.loads(json_text)
            return data['edi_expression']
        except Exception as e:
            # Fallback parsing
            return "Error"

# Chain to generate EDI expression given the segment and entities extracted
prompt_template_edi_expression = ChatPromptTemplate.from_template("""
You are an expert EDI (Electronic Data Interchange) data extraction specialist. Your task is to generate an EDI expression for the segment and the entities extracted below.
Version: 006010

Segment: {segment}
Entities: {entities}

Output format:
```
{{
    "edi_expression": "EDI_EXPRESSION"
}}
```
""")

# Create the chain
chain_edi_expression = (
    prompt_template_edi_expression
    | llm
    | EDIExpressionOutputParser()
)
````
